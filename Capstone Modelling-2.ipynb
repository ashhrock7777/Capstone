```python
import json

def load_processed_data(file_path):
    with open(file_path,'r') as f:
        data = json.load(f)
    return data

json_file_path = 'processed_midi_data.json'

processed_data = load_processed_data(json_file_path)
```

```python
import pandas as pd

json_df = pd.json_normalize(processed_data)

json_df['filename'] = json_df['filename'].str.replace(r'.mid','',regex=True)
```
